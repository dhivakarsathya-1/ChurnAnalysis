create a new azure sql database and create the following control metadata table which we will be using for raw data ingestion

### Control Metadata table

```python
%sql

create table [dbo].[metadata_control](
    id INT PRIMARY KEY,
	groupId INT,
    tableName NVARCHAR(255),
    sourcePath NVARCHAR(1000),
    fileFormat NVARCHAR(50),
    schemaName NVARCHAR(100),
    isActive BIT DEFAULT 1,
    lastUpdatedAt DATETIME2,
    updatedBy NVARCHAR(100),
    dqChecks NVARCHAR(MAX),
    checkpointLocation NVARCHAR(1000),
    description NVARCHAR(500)
);
```
